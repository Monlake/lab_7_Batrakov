```python
import heapq
import random
class MaxHeap:
    def __init__(self):
        self.heap = []
    def heappush(self, item):
        heapq.heappush(self.heap, -item)
    def heappop(self):
        return -heapq.heappop(self.heap)
    def __str__(self):
        return str([-x for x in self.heap])
def generate_heap_from_array(array):
    max_heap = MaxHeap()
    for num in array:
        max_heap.heappush(num)
    return max_heap
def add_to_heap(heap, element):
    heap.heappush(element)
    return heap
def remove_max(heap):
    return heap.heappop()
random_array = random.sample(range(1, 100), 10)
print("Рандомний масив:", random_array)
max_heap = generate_heap_from_array(random_array)
print("Максимальна купа з рандомного масиву:", max_heap)
my_number = 2
add_to_heap(max_heap, my_number)
print("Максимальна купа після додавання елемента", my_number, ":", max_heap)
max_element = remove_max(max_heap)
print("Максимальний елемент, вилучений з купи:", max_element)
print("Максимальна купа після вилучення максимального елемента:", max_heap)
print("""
Оцінка складності:
Search (пошук): у середньому і в найгіршому випадках - O(n)
Insert (вставка): у середньому і в найгіршому випадках - O(log n)
Delete (видалення): у середньому і в найгіршому випадках - O(log n)""")
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * self.size
    
    def hash_function(self, key):
        return key % self.size
    
    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index] = value
    
    def search(self, key):
        index = self.hash_function(key)
        return self.table[index]
    
    def delete(self, key):
        index = self.hash_function(key)
        self.table[index] = None
```

```python
import time

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(self.size)]
    
    def hash_function(self, key):
        if isinstance(key, int):
            return key % self.size
        elif isinstance(key, str):
            return sum(ord(char) for char in key) % self.size
        elif isinstance(key, tuple):
            return sum(hash(elem) for elem in key) % self.size
        elif isinstance(key, dict):
            return sum(hash(key) for key in key.items()) % self.size
        elif hasattr(key, '__hash__'):
            return hash(key) % self.size
        else:
            raise TypeError("Unsupported key type")
    
    def insert(self, key, value):
        index = self.hash_function(key)
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value
                return
        self.table[index].append([key, value])
    
    def search(self, key):
        index = self.hash_function(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return pair[1]
        return None
    
    def delete(self, key):
        index = self.hash_function(key)
        for pair in self.table[index]:
            if pair[0] == key:
                self.table[index].remove(pair)
                return

def test_hash_table(data_type, data):
    hash_table = HashTable(100)
    
    start_time = time.time()
    for key, value in data.items():
        hash_table.insert(key, value)
    insert_time = time.time() - start_time
    
    start_time = time.time()
    for key in data:
        hash_table.search(key)
    search_time = time.time() - start_time
    
    start_time = time.time()
    for key in data:
        hash_table.delete(key)
    delete_time = time.time() - start_time
    
    print(f"Тип даних: {data_type}")
    print(f"Час вставки: {insert_time:.6f} сек")
    print(f"Час пошуку: {search_time:.6f} сек")
    print(f"Час видалення: {delete_time:.6f} сек")
    print()

if __name__ == "__main__":
    int_data = {i: f"value_{i}" for i in range(100)}
    test_hash_table("цілі числа", int_data)
    
    str_data = {str(i): f"value_{i}" for i in range(100)}
    test_hash_table("рядки", str_data)
    
    list_data = {tuple(range(i)): f"value_{i}" for i in range(100)}
    test_hash_table("списки", list_data)
    
    dict_data = {f"key_{i}": {"value": f"value_{i}"} for i in range(100)}
    test_hash_table("словники", dict_data)
    
    class Obj:
        def __init__(self, value):
            self.value = value
        
        def __hash__(self):
            return hash(self.value)
    
    obj_data = {Obj(i): f"value_{i}" for i in range(100)}
    test_hash_table("об'єкти", obj_data)
```
